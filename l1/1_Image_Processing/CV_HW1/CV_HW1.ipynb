```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np
import imageio
import cv2

%matplotlib inline
plt.rcParams["figure.figsize"] = (8, 10)
```

```python
image = cv2.imread('Loki.jpg')
image = cv2.resize(image, (int(image.shape[1]/5), int(image.shape[0]/5)))
plt.imshow(image)
plt.show()
```

```python
# Перевод в нормальную цветовую шкалу, по умолчанию в OpenCV стоит BGR
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
plt.imshow(RGB_img)
plt.show()
```

```python
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
plt.imshow(RGB_img, cmap='gray')
plt.show()
```

```python
lab_img = cv2.cvtColor(image, cv2.cv2.COLOR_BGR2LAB)
plt.imshow(lab_img)
plt.show()
```

```python
# С hsv муть какая-то на картинке
hsv_img = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_img)
plt.show()
```

```python
# Блюры разной степени и разного типа
plt.figure(figsize=(22, 10))

plt.subplot(131)
blurred_img = cv2.GaussianBlur(image, (51, 51), 0)
plt.imshow(blurred_img)

plt.subplot(132)
blurred_img = cv2.medianBlur(image, 9)
plt.imshow(blurred_img)

plt.subplot(133)
blurred_img = cv2.bilateralFilter(image, 9, 75, 75)
plt.imshow(blurred_img)

plt.show()
```

```python
# Контурирование
original_image = Image.open('Loki.jpg')
original_image = original_image.resize((int(image.shape[1]/5), int(image.shape[0]/5)))
contoured_image = original_image.filter(ImageFilter.CONTOUR)
plt.imshow(contoured_image)
plt.show()
```

```python
# Повышение резкости разной степени
plt.figure(figsize=(22, 10))

plt.subplot(131)
detailed_image = original_image.filter(ImageFilter.DETAIL)
plt.imshow(detailed_image)

plt.subplot(132)
enhanced_image = original_image.filter(ImageFilter.EDGE_ENHANCE)
plt.imshow(enhanced_image)

plt.subplot(133)
enhanced2_image = original_image.filter(ImageFilter.EDGE_ENHANCE_MORE)
plt.imshow(enhanced2_image)

plt.show()
```

```python
# Эффект штамповки
embossed_image = original_image.filter(ImageFilter.EMBOSS)
plt.imshow(embossed_image)
plt.show()
```

```python
# Эффект выделения границ
find_edged_image = original_image.filter(ImageFilter.FIND_EDGES)
plt.imshow(find_edged_image)
plt.show()
```
